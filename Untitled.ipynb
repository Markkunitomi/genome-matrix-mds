```python
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter

from sklearn import manifold, datasets

n_points = 284
n_neighbors = 30
n_components = 2

def Create_SE_plot(Distance_matrix, INDEX):
    se = manifold.SpectralEmbedding(n_components=n_components,
                                n_neighbors=n_neighbors,
                                affinity="precomputed")
    results = se.fit(Distance_matrix)
    coords = results.embedding_
    Plot = pd.DataFrame(dict(x=coords[:, 0], y=coords[:, 1], label=INDEX))
    plt.rcParams["figure.figsize"] = fig_size
    groups = Plot.groupby('label')
    fig, ax = plt.subplots()
    ax.margins(0.2) # Optional, just adds 5% padding to the autoscaling
    
    colors = cm.tab10(np.linspace(0, 1, len(groups)))
    
    for group, color in zip(groups, colors):
        ax.plot(group[1].x, group[1].y, marker='o', linestyle='', ms=2, label=group[0], c=color)
        

    lgd = pylab.legend()
    ax.legend_.remove()
    plt.show()
```

```python
from sklearn import manifold

n_points = 284
n_neighbors = 20
n_components = 2

def Create_IsoMap_plot(Distance_matrix, INDEX):
    IsoMap = manifold.Isomap(n_neighbors, n_components).fit(Distance_matrix)
#     print(Distance_matrix)
    results = IsoMap.fit(Distance_matrix)
    coords = results.embedding_
    Plot = pd.DataFrame(dict(x=coords[:, 0], y=coords[:, 1], label=INDEX))
    plt.rcParams["figure.figsize"] = fig_size
    groups = Plot.groupby('label')
    fig, ax = plt.subplots()
    ax.margins(0.2) # Optional, just adds 5% padding to the autoscaling
    
    colors = cm.tab10(np.linspace(0, 1, len(groups)))
    
    for group, color in zip(groups, colors):
        ax.plot(group[1].x, group[1].y, marker='o', linestyle='', ms=2, label=group[0], c=color)

    lgd = pylab.legend()
    ax.legend_.remove()
    plt.show()
```

```python
### Manifold transformation


n_points = 284
n_neighbors = 30
n_components = 2

methods = ['standard', 'ltsa', 'hessian', 'modified']
labels = ['LLE', 'LTSA', 'Hessian LLE', 'Modified LLE']      
        
def Create_Manifold_plot(Distance_matrix, INDEX):
    for i, method in enumerate(methods):
        embedding = manifold.LocallyLinearEmbedding(n_neighbors, n_components,
                                            eigen_solver='auto',
                                            method=method).fit(Distance_matrix)
        results = embedding.fit(Distance_matrix)
        coords = results.embedding_
        Plot = pd.DataFrame(dict(x=coords[:, 0], y=coords[:, 1], label=INDEX))

        plt.rcParams["figure.figsize"] = fig_size
        groups = Plot.groupby('label')
        fig, ax = plt.subplots()
        ax.margins(0.2) # Optional, just adds 5% padding to the autoscaling

        for group, color in zip(groups, colors):
            ax.plot(group[1].x, group[1].y, marker='o', linestyle='', ms=2, label=group[0], c=color)


        lgd = pylab.legend()
        ax.legend_.remove()
        plt.show()
```

```python
import gzip

with gzip.open("GenBank_Bacterial_taxID_lineage.tsv.gz", "rt") as file:
    GB_Lines = [Entry.split("\t") for Entry in file.readlines()]

with gzip.open("IBM_assemblies_taxonomy_metadata.tsv.gz", "rt") as file:
    IBM_Lines = [Entry.split("\t") for Entry in file.readlines()]
```

```python
Name_to_Genus = {}
Name_to_Species = {}
Name_to_taxID = {}

### GenBank
for Line in GB_Lines[1:]:
    try:
        File = Line[0].split(".")[0]

        taxID = Line[1]

        Rank_lineage = Line[4].split(",")
        Name_lineage = Line[2].split(",")
        Species_Index = Rank_lineage.index("species")
        Species = Name_lineage[Species_Index]
        Genus_Index = Rank_lineage.index("genus")
        Genus = Name_lineage[Genus_Index]
#         print(File, taxID, Species, Genus)
        Name_to_Genus[File] = Genus
        Name_to_Species[File] = Species
        Name_to_taxID[File] = taxID
    except:
        print(Line)

        continue
        
### IBM
for Line in IBM_Lines[1:]:
    try:
        File = Line[0]
        
        taxID = Line[1]

        Rank_lineage = Line[4].split(",")
        Name_lineage = Line[2].split(",")
        Species_Index = Rank_lineage.index("species")
        Species = Name_lineage[Species_Index]
        Genus_Index = Rank_lineage.index("genus")
        Genus = Name_lineage[Genus_Index]
#         print(File, taxID, Species, Genus)
        Name_to_Genus[File] = Genus
        Name_to_Species[File] = Species
        Name_to_taxID[File] = taxID
    except:
#         print(Line)
        continue
        
Label=[]
Genus = []
Species = []
taxID = []
for entry in list(df):
    try:
        
        entry = entry.split("/")[-1]
        if entry.startswith("GCA"):
            entry = entry.split(".")[0]
        if "-" in entry:
            entry = entry.split("-")[0]
        Label.append(entry)
        taxID.append(Name_to_taxID[entry])
        Species.append(Name_to_Species[entry])
        Genus.append(Name_to_Genus[entry])
        
    except:
        print(entry)
        taxID.append("NA")
        Species.append("NA")
        Genus.append("NA")
```
